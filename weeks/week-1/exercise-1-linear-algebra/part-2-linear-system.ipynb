# Part 2 - Solving a Linear System

This notebook solves the second part of Exercise 1 step by step. The goal is to rewrite the system in matrix form `A @ x = b`, then solve it with the inverse matrix method `x = A^-1 @ b`.

## 1. Original system

We are given three equations with three unknowns:

```text
x1 + x2 + x3 = 6
x1 - x2 + x3 = 2
x1 + x2 - x3 = 0
```

The numbers multiplying `x1`, `x2`, and `x3` are the coefficients. These coefficients will become the rows of matrix `A`.

## 2. Matrix form

The matrix equation is:

```text
A @ x = b
```

where:

```text
A = [[1,  1,  1],
     [1, -1,  1],
     [1,  1, -1]]

x = [[x1],
     [x2],
     [x3]]

b = [[6],
     [2],
     [0]]
```

Reasoning: each row of `A` matches one equation. For example, the second row `[1, -1, 1]` means `x1 - x2 + x3`.

````python
import numpy as np

A = np.array([
    [1,  1,  1],
    [1, -1,  1],
    [1,  1, -1],
], dtype=float)

b = np.array([
    [6],
    [2],
    [0],
], dtype=float)

print('A =')
print(A)
print('\nb =')
print(b)
````

## 3. Check that A is invertible

The inverse method only works if `A` is invertible. A square matrix is invertible when its determinant is not zero.

````python
det_A = np.linalg.det(A)
print(f'det(A) = {det_A:.0f}')

if abs(det_A) < 1e-12:
    raise ValueError('A is not invertible, so the inverse method cannot be used.')

print('A is invertible because det(A) is not zero.')
````

## 4. Compute the inverse and solve

Because `A` is invertible, multiply both sides of `A @ x = b` by `A^-1`:

```text
A^-1 @ A @ x = A^-1 @ b
x = A^-1 @ b
```

````python
A_inverse = np.linalg.inv(A)
x = A_inverse @ b

print('A inverse =')
print(A_inverse)

print('\nSolution vector x =')
print(x)
````

## 5. Final answer

The solution vector is:

```text
x = [[1],
     [2],
     [3]]
```

So:

```text
x1 = 1
x2 = 2
x3 = 3
```

## 6. Verify the answer

A good solution should be checked by substituting the values back into the original equations. In matrix form, this means checking that `A @ x` equals `b`.

````python
left_side = A @ x

print('A @ x =')
print(left_side)

print('\nb =')
print(b)

assert np.allclose(left_side, b)
print('\nVerification passed: A @ x equals b.')
````

## 7. Short written explanation

The matrix `A` contains the coefficients from the three equations, the vector `x` contains the unknown variables, and the vector `b` contains the constants on the right side. Since `det(A) = 4`, `A` is invertible. Therefore the equation `A @ x = b` can be solved by multiplying `b` by `A^-1`. The result is `x = [1, 2, 3]^T`, which satisfies all three original equations.
